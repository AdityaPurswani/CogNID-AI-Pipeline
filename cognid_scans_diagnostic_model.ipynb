```python
import pandas as pd
import numpy as np

# Load the data
file_path = 'cognid_scans_filtered.csv'
features_df = pd.read_csv(file_path)
print(f"Loaded dataset with {features_df.shape[0]} rows and {features_df.shape[1]} columns")

# Get all column names
columns = features_df.columns.tolist()

# Filter out columns with Surface_Area or Eccentricity - these will be excluded
columns_to_use = [col for col in columns if 'Surface_Area' not in col and 'Eccentricity' not in col]

# Create a new DataFrame with filtered columns
filtered_df = features_df[columns_to_use].copy()
print(f"After removing Surface_Area and Eccentricity: {filtered_df.shape[1]} columns")

# ------------------ HIPPOCAMPUS ------------------
# Check if hippocampus columns exist
if 'left_hippocampus_Volume_mm3' in columns and 'right_hippocampus_Volume_mm3' in columns:
    # Create average volume
    filtered_df['Hippocampus_Volume'] = (filtered_df['left_hippocampus_Volume_mm3'] + filtered_df['right_hippocampus_Volume_mm3']) / 2
    
    # Create volume ratio
    filtered_df['Hippocampus_Volume_Ratio'] = (filtered_df['left_hippocampus_Volume_mm3'] / filtered_df['right_hippocampus_Volume_mm3'] )

# Check if hippocampus compactness columns exist
if 'left_hippocampus_Compactness' in columns and 'right_hippocampus_Compactness' in columns:
    # Create average compactness
    filtered_df['Hippocampus_Compactness'] = (filtered_df['left_hippocampus_Compactness'] + filtered_df['right_hippocampus_Compactness']) / 2
    
    # Create compactness ratio
    filtered_df['Hippocampus_Compactness_Ratio'] = (filtered_df['left_hippocampus_Compactness'] / filtered_df['right_hippocampus_Compactness'] )

# Check if hippocampus sphericity columns exist
if 'left_hippocampus_Sphericity' in columns and 'right_hippocampus_Sphericity' in columns:
    # Create average sphericity
    filtered_df['Hippocampus_Sphericity'] = (filtered_df['left_hippocampus_Sphericity'] + filtered_df['right_hippocampus_Sphericity']) / 2
    
    # Create sphericity ratio
    filtered_df['Hippocampus_Sphericity_Ratio'] = (filtered_df['left_hippocampus_Sphericity'] / filtered_df['right_hippocampus_Sphericity'])

# ------------------ AMYGDALA ------------------
# Check if amygdala columns exist
if 'left_amygdala_Volume_mm3' in columns and 'right_amygdala_Volume_mm3' in columns:
    # Create average volume
    filtered_df['Amygdala_Volume'] = (filtered_df['left_amygdala_Volume_mm3'] + filtered_df['right_amygdala_Volume_mm3']) / 2
    
    # Create volume ratio
    filtered_df['Amygdala_Volume_Ratio'] = (filtered_df['left_amygdala_Volume_mm3'] / filtered_df['right_amygdala_Volume_mm3'])

# Check if amygdala compactness columns exist
if 'left_amygdala_Compactness' in columns and 'right_amygdala_Compactness' in columns:
    # Create average compactness
    filtered_df['Amygdala_Compactness'] = (filtered_df['left_amygdala_Compactness'] + filtered_df['right_amygdala_Compactness']) / 2
    
    # Create compactness ratio
    filtered_df['Amygdala_Compactness_Ratio'] = (filtered_df['left_amygdala_Compactness'] / filtered_df['right_amygdala_Compactness'])

# Check if amygdala sphericity columns exist
if 'left_amygdala_Sphericity' in columns and 'right_amygdala_Sphericity' in columns:
    # Create average sphericity
    filtered_df['Amygdala_Sphericity'] = (filtered_df['left_amygdala_Sphericity'] + filtered_df['right_amygdala_Sphericity']) / 2
    
    # Create sphericity ratio
    filtered_df['Amygdala_Sphericity_Ratio'] = (filtered_df['left_amygdala_Sphericity'] / filtered_df['right_amygdala_Sphericity'])

# ------------------ LATERAL VENTRICLE ------------------
# Check if lateral ventricle columns exist
if 'left_lateral_ventricle_Volume_mm3' in columns and 'right_lateral_ventricle_Volume_mm3' in columns:
    # Create average volume
    filtered_df['Lateral_ventricle_Volume'] = (filtered_df['left_lateral_ventricle_Volume_mm3'] + filtered_df['right_lateral_ventricle_Volume_mm3']) / 2
    
    # Create volume ratio
    filtered_df['Lateral_ventricle_Volume_Ratio'] = (filtered_df['left_lateral_ventricle_Volume_mm3'] / filtered_df['right_lateral_ventricle_Volume_mm3'])

# Check if lateral ventricle compactness columns exist
if 'left_lateral_ventricle_Compactness' in columns and 'right_lateral_ventricle_Compactness' in columns:
    # Create average compactness
    filtered_df['Lateral_ventricle_Compactness'] = (filtered_df['left_lateral_ventricle_Compactness'] + filtered_df['right_lateral_ventricle_Compactness']) / 2
    
    # Create compactness ratio
    filtered_df['Lateral_ventricle_Compactness_Ratio'] = (filtered_df['left_lateral_ventricle_Compactness'] / filtered_df['right_lateral_ventricle_Compactness'])

# Check if lateral ventricle sphericity columns exist
if 'left_lateral_ventricle_Sphericity' in columns and 'right_lateral_ventricle_Sphericity' in columns:
    # Create average sphericity
    filtered_df['Lateral_ventricle_Sphericity'] = (filtered_df['left_lateral_ventricle_Sphericity'] + filtered_df['right_lateral_ventricle_Sphericity']) / 2
    
    # Create sphericity ratio
    filtered_df['Lateral_ventricle_Sphericity_Ratio'] = (filtered_df['left_lateral_ventricle_Sphericity'] / filtered_df['right_lateral_ventricle_Sphericity'])

# ------------------ CORTEX ------------------
# Check if cortex columns exist
if 'left_cortex_Volume_mm3' in columns and 'right_cortex_Volume_mm3' in columns:
    # Create average volume
    filtered_df['Cortex_Volume'] = (filtered_df['left_cortex_Volume_mm3'] + filtered_df['right_cortex_Volume_mm3']) / 2
    
    # Create volume ratio
    filtered_df['Cortex_Volume_Ratio'] = (filtered_df['left_cortex_Volume_mm3'] / filtered_df['right_cortex_Volume_mm3'])

# Check if cortex compactness columns exist
if 'left_cortex_Compactness' in columns and 'right_cortex_Compactness' in columns:
    # Create average compactness
    filtered_df['Cortex_Compactness'] = (filtered_df['left_cortex_Compactness'] + filtered_df['right_cortex_Compactness']) / 2
    
    # Create compactness ratio
    filtered_df['Cortex_Compactness_Ratio'] = (filtered_df['left_cortex_Compactness'] / filtered_df['right_cortex_Compactness'])

# Check if cortex sphericity columns exist
if 'left_cortex_Sphericity' in columns and 'right_cortex_Sphericity' in columns:
    # Create average sphericity
    filtered_df['Cortex_Sphericity'] = (filtered_df['left_cortex_Sphericity'] + filtered_df['right_cortex_Sphericity']) / 2
    
    # Create sphericity ratio
    filtered_df['Cortex_Sphericity_Ratio'] = (filtered_df['left_cortex_Sphericity'] / filtered_df['right_cortex_Sphericity'])

# ------------------ CAUDATE ------------------
# Check if caudate columns exist
if 'left_caudate_Volume_mm3' in columns and 'right_caudate_Volume_mm3' in columns:
    # Create average volume
    filtered_df['Caudate_Volume'] = (filtered_df['left_caudate_Volume_mm3'] + filtered_df['right_caudate_Volume_mm3']) / 2
    
    # Create volume ratio
    filtered_df['Caudate_Volume_Ratio'] = (filtered_df['left_caudate_Volume_mm3'] / filtered_df['right_caudate_Volume_mm3'])

# Check if caudate compactness columns exist
if 'left_caudate_Compactness' in columns and 'right_caudate_Compactness' in columns:
    # Create average compactness
    filtered_df['Caudate_Compactness'] = (filtered_df['left_caudate_Compactness'] + filtered_df['right_caudate_Compactness']) / 2
    
    # Create compactness ratio
    filtered_df['Caudate_Compactness_Ratio'] = (filtered_df['left_caudate_Compactness'] / filtered_df['right_caudate_Compactness'])

# Check if caudate sphericity columns exist
if 'left_caudate_Sphericity' in columns and 'right_caudate_Sphericity' in columns:
    # Create average sphericity
    filtered_df['Caudate_Sphericity'] = (filtered_df['left_caudate_Sphericity'] + filtered_df['right_caudate_Sphericity']) / 2
    
    # Create sphericity ratio
    filtered_df['Caudate_Sphericity_Ratio'] = (filtered_df['left_caudate_Sphericity'] / filtered_df['right_caudate_Sphericity'])

# ------------------ PUTAMEN ------------------
# Check if putamen columns exist
if 'left_putamen_Volume_mm3' in columns and 'right_putamen_Volume_mm3' in columns:
    # Create average volume
    filtered_df['Putamen_Volume'] = (filtered_df['left_putamen_Volume_mm3'] + filtered_df['right_putamen_Volume_mm3']) / 2
    
    # Create volume ratio
    filtered_df['Putamen_Volume_Ratio'] = (filtered_df['left_putamen_Volume_mm3'] / filtered_df['right_putamen_Volume_mm3'])

# Check if putamen compactness columns exist
if 'left_putamen_Compactness' in columns and 'right_putamen_Compactness' in columns:
    # Create average compactness
    filtered_df['Putamen_Compactness'] = (filtered_df['left_putamen_Compactness'] + filtered_df['right_putamen_Compactness']) / 2
    
    # Create compactness ratio
    filtered_df['Putamen_Compactness_Ratio'] = (filtered_df['left_putamen_Compactness'] / filtered_df['right_putamen_Compactness'])

# Check if putamen sphericity columns exist
if 'left_putamen_Sphericity' in columns and 'right_putamen_Sphericity' in columns:
    # Create average sphericity
    filtered_df['Putamen_Sphericity'] = (filtered_df['left_putamen_Sphericity'] + filtered_df['right_putamen_Sphericity']) / 2
    
    # Create sphericity ratio
    filtered_df['Putamen_Sphericity_Ratio'] = (filtered_df['left_putamen_Sphericity'] / filtered_df['right_putamen_Sphericity']
    )

# ------------------ THALAMUS ------------------
# Check if thalamus columns exist
if 'left_thalamus_Volume_mm3' in columns and 'right_thalamus_Volume_mm3' in columns:
    # Create average volume
    filtered_df['Thalamus_Volume'] = (filtered_df['left_thalamus_Volume_mm3'] + filtered_df['right_thalamus_Volume_mm3']) / 2
    
    # Create volume ratio
    filtered_df['Thalamus_Volume_Ratio'] = (filtered_df['left_thalamus_Volume_mm3'] / filtered_df['right_thalamus_Volume_mm3'])

# Check if thalamus compactness columns exist
if 'left_thalamus_Compactness' in columns and 'right_thalamus_Compactness' in columns:
    # Create average compactness
    filtered_df['Thalamus_Compactness'] = (filtered_df['left_thalamus_Compactness'] + filtered_df['right_thalamus_Compactness']) / 2
    
    # Create compactness ratio
    filtered_df['Thalamus_Compactness_Ratio'] = (filtered_df['left_thalamus_Compactness'] / filtered_df['right_thalamus_Compactness'])

# Check if thalamus sphericity columns exist
if 'left_thalamus_Sphericity' in columns and 'right_thalamus_Sphericity' in columns:
    # Create average sphericity
    filtered_df['Thalamus_Sphericity'] = (filtered_df['left_thalamus_Sphericity'] + filtered_df['right_thalamus_Sphericity']) / 2
    
    # Create sphericity ratio
    filtered_df['Thalamus_Sphericity_Ratio'] = (filtered_df['left_thalamus_Sphericity'] / filtered_df['right_thalamus_Sphericity'])

# ------------------ PARAHIPPOCAMPAL GYRUS ANTERIOR/POSTERIOR ------------------
# Check if parahippocampal_gyrus columns exist
ant_col = 'parahippocampal_gyrus_anterior_Volume_mm3'
post_col = 'parahippocampal_gyrus_posterior_Volume_mm3'
if ant_col in columns and post_col in columns:
    # Create total volume
    filtered_df['Parahippocampal_gyrus_Volume'] = (filtered_df[ant_col] + filtered_df[post_col]) / 2
    
    # Create anterior/posterior ratio
    filtered_df['Parahippocampal_gyrus_AP_Volume_Ratio'] = (filtered_df[ant_col] / filtered_df[post_col])

# Check if parahippocampal_gyrus compactness columns exist
ant_col = 'parahippocampal_gyrus_anterior_Compactness'
post_col = 'parahippocampal_gyrus_posterior_Compactness'
if ant_col in columns and post_col in columns:
    # Create average compactness
    filtered_df['Parahippocampal_gyrus_Compactness'] = (filtered_df[ant_col] + filtered_df[post_col]) / 2
    
    # Create anterior/posterior ratio
    filtered_df['Parahippocampal_gyrus_AP_Compactness_Ratio'] = (filtered_df[ant_col] / filtered_df[post_col])

# Check if parahippocampal_gyrus sphericity columns exist
ant_col = 'parahippocampal_gyrus_anterior_Sphericity'
post_col = 'parahippocampal_gyrus_posterior_Sphericity'
if ant_col in columns and post_col in columns:
    # Create average sphericity
    filtered_df['Parahippocampal_gyrus_Sphericity'] = (filtered_df[ant_col] + filtered_df[post_col]) / 2
    
    # Create anterior/posterior ratio
    filtered_df['Parahippocampal_gyrus_AP_Sphericity_Ratio'] = (filtered_df[ant_col] / filtered_df[post_col])

# After all the feature engineering, create final column selection

# Identify all engineered features
original_columns = set(columns)
all_columns = set(filtered_df.columns)
engineered_features = list(all_columns - original_columns)

# Create filters for columns to exclude
exclude_patterns = [
    'left_', 'right_',                        # Left/right pairs               
    'parahippocampal_gyrus_anterior',         # Special case with anterior in name
    'parahippocampal_gyrus_posterior'         # Special case with posterior in name
]

# Filter the columns for final output
final_columns = []

# Always include MRI_ID
if 'MRI_ID' in columns:
    final_columns.append('MRI_ID')

# Add all engineered features
final_columns.extend(engineered_features)

# Include region features that aren't part of pairs
for col in columns:
    # Check if it's MRI_ID or Completed Diagnosis - already handled separately
    if col in ['MRI_ID', 'Completed Diagnosis']:
        continue
        
    # Check if Surface_Area or Eccentricity - should be excluded
    if 'Surface_Area' in col or 'Eccentricity' in col:
        continue
        
    # Check if it's one of the excluded patterns
    exclude = False
    for pattern in exclude_patterns:
        if pattern in col:
            exclude = True
            break
            
    # If not excluded, include it
    if not exclude:
        final_columns.append(col)

# Add Completed Diagnosis if it exists
if 'Completed Diagnosis' in columns:
    print(True)
    final_columns.append('Completed Diagnosis')
else:
    print(False)

# Create the final dataframe with selected columns
final_df = filtered_df[final_columns]

# Save the final dataset to CSV
output_path = 'cognid_scans_engineered_features.xlsx'
final_df.to_excel(output_path, index=False)

# Count column types
paired_excluded = len([col for col in columns if any(pattern in col for pattern in exclude_patterns)])
original_kept = len([col for col in final_columns if col in columns and col != 'MRI_ID' and col != 'Completed Diagnosis'])
engineered_count = len(engineered_features)

print(f"Final dataset statistics:")
print(f"  - {len(final_columns)} total columns")
print(f"  - {engineered_count} engineered features")
print(f"  - {original_kept} original features kept")
print(f"  - {paired_excluded} paired region features excluded")
print(f"  - Plus ID and diagnosis columns (if present)")

# Print sample of the final columns
print("\nSample of final columns:")
sample_size = min(100, len(final_columns))
print(final_columns[:sample_size])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest, f_classif, RFE, mutual_info_classif, chi2
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix

# Load the data
df = pd.read_excel('cognid_scans_engineered_features.xlsx')

# Check basic information about the dataset
print("Dataset shape:", df.shape)
print("\nTarget variable distribution:")
print(df['Completed Diagnosis'].value_counts())

# Check for missing values
missing_values = df.isnull().sum()
print("\nColumns with missing values:")
print(missing_values[missing_values > 0])

# Handle missing values in the target
df['Completed Diagnosis'] = df['Completed Diagnosis'].fillna('No diagnosis')

# df = df[df['Completed Diagnosis'] != 'No diagnosis']

# Remove rows with "No diagnosis"
print(f"\nOriginal dataset shape: {df.shape}")
df = df[df['Completed Diagnosis'] != 'No diagnosis']
print(f"Dataset shape after removing 'No diagnosis' rows: {df.shape}")

df.to_excel("cognid_scans_cleaned.xlsx")

# Convert categorical target to numerical using label encoding
le = LabelEncoder()
y = le.fit_transform(df['Completed Diagnosis'])

# Map numerical labels back to original categories for reference
label_mapping = dict(zip(range(len(le.classes_)), le.classes_))
print("\nTarget Label Mapping:")
for key, value in label_mapping.items():
    print(f"{key}: {value}")

# Create feature matrix (exclude the MRI_ID and target column)
X = df.drop(['MRI_ID', 'Completed Diagnosis'], axis=1)

# Check for and handle any remaining missing values in features
if X.isnull().sum().sum() > 0:
    X = X.fillna(X.mean())

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42, stratify=y
)

print(f"\nTraining set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
# 1. Filter Method: ANOVA F-value
print("\n---------- ANOVA F-VALUE FEATURE SELECTION ----------")
k_best = 25# Number of features to select
selector_f = SelectKBest(f_classif, k=k_best)
X_kbest = selector_f.fit_transform(X_scaled, y)

# Get selected feature names
feature_indices_f = selector_f.get_support(indices=True)
selected_features_f = X.columns[feature_indices_f]
print(f"Top {k_best} features selected by ANOVA F-test:")
for i, feature in enumerate(selected_features_f):
    score = selector_f.scores_[feature_indices_f[i]]
    print(f"{feature}: {score:.4f}")

# 2. Filter Method: Mutual Information (for categorical targets)
print("\n---------- MUTUAL INFORMATION FEATURE SELECTION ----------")
selector_mi = SelectKBest(mutual_info_classif, k=k_best)
X_mi = selector_mi.fit_transform(X_scaled, y)

# Get selected feature names
feature_indices_mi = selector_mi.get_support(indices=True)
selected_features_mi = X.columns[feature_indices_mi]
print(f"Top {k_best} features selected by Mutual Information:")
for i, feature in enumerate(selected_features_mi):
    score = selector_mi.scores_[feature_indices_mi[i]]
    print(f"{feature}: {score:.4f}")

# 3. Wrapper Method: Recursive Feature Elimination with RandomForest
print("\n---------- RECURSIVE FEATURE ELIMINATION ----------")
model_rfe = RandomForestClassifier(n_estimators=100, random_state=42)
rfe = RFE(estimator=model_rfe, n_features_to_select=k_best, step=1)
rfe.fit(X_scaled, y)

# Get selected feature names
selected_features_rfe = X.columns[rfe.support_]
print(f"Top {k_best} features selected by RFE with RandomForest:")
for feature in selected_features_rfe:
    print(f"{feature}: {rfe.ranking_[list(X.columns).index(feature)]}")

# 4. Embedded Method: Random Forest Feature Importance
print("\n---------- RANDOM FOREST FEATURE IMPORTANCE ----------")
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_scaled, y)

# Get feature importances
importances = model_rf.feature_importances_
indices = np.argsort(importances)[::-1]

print(f"Top {k_best} features selected by Random Forest Feature Importance:")
top_rf_features = []
for i in range(min(k_best, len(X.columns))):
    feature = X.columns[indices[i]]
    importance = importances[indices[i]]
    top_rf_features.append(feature)
    print(f"{feature}: {importance:.4f}")
```

```python
print("\n---------- FEATURE SELECTION OVERLAP ANALYSIS ----------")
all_selected = set(selected_features_f) | set(selected_features_mi) | set(selected_features_rfe) | set(top_rf_features)
common_features = set(selected_features_f) & set(selected_features_mi) & set(selected_features_rfe) & set(top_rf_features)

print(f"Total number of unique selected features across all methods: {len(all_selected)}")
print(f"Number of common features selected by all methods: {len(common_features)}")

if len(common_features) > 0:
    print("Common features across all methods:")
    for feature in common_features:
        print(f"- {feature}")

# Create a set of features that appear in at least 3 of the 4 methods
features_in_three_methods = set()
for feature in all_selected:
    count = 0
    if feature in selected_features_f:
        count += 1
    if feature in selected_features_mi:
        count += 1
    if feature in selected_features_rfe:
        count += 1
    if feature in top_rf_features:
        count += 1
    
    if count >= 3:
        features_in_three_methods.add(feature)

print(f"\nFeatures selected by at least 3 methods: {len(features_in_three_methods)}")
for feature in features_in_three_methods:
    print(f"- {feature}")
```

```python
print("\n---------- MULTICOLLINEARITY ANALYSIS ----------")
# Calculate correlation matrix for original features
corr_matrix = X.corr().abs()
selected_features = list(selected_features_rfe)
# Additional features for classes with less instances
additional_features = ['superior_parietal_lobule_Volume_mm3', 'parietal_lobe_Volume_mm3', 'superior_frontal_gyrus_Volume_mm3', 'middle_frontal_gyrus_Volume_mm3', 'frontal_orbital_cortex_Volume_mm3', 'frontal_lobe_Volume_mm3']
selected_features.extend(additional_features)
# Identify highly correlated features
threshold = 0.9
high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > threshold:
            col_i = corr_matrix.columns[i]
            col_j = corr_matrix.columns[j]
            corr_value = corr_matrix.iloc[i, j]
            high_corr_pairs.append((col_i, col_j, corr_value))

print(f"Number of highly correlated feature pairs (r > {threshold}): {len(high_corr_pairs)}")
if len(high_corr_pairs) > 0:
    print("Top 10 highest correlated pairs:")
    high_corr_pairs.sort(key=lambda x: x[2], reverse=True)
    for i, (col_i, col_j, corr_value) in enumerate(high_corr_pairs[:10]):
        print(f"{col_i} - {col_j}: {corr_value:.4f}")

# Save correlation heatmap of selected features (optional)
# Comment out if you don't need the visualization
try:
    plt.figure(figsize=(15, 12))
    selected_features_list = list(selected_features)
    selected_corr = df[selected_features_list].corr().abs()
    sns.heatmap(selected_corr, annot=True, cmap='coolwarm', fmt='.2f', 
                xticklabels=selected_features_list, yticklabels=selected_features_list)
    plt.title('Correlation Matrix of Selected Features')
    plt.tight_layout()
    plt.savefig('selected_features_correlation.png')
    print("Correlation heatmap saved as 'selected_features_correlation.png'")
except Exception as e:
    print(f"Could not generate correlation heatmap: {e}")
```

```python
print("\n---------- GROUP-BASED FEATURE SELECTION ----------")
# Group features by brain region
region_groups = {}
for col in X.columns:
    parts = col.split('_')
    if len(parts) >= 2:
        # Extract brain region by removing the measurement type and unit
        if parts[-1] in ['mm3', 'mm2']:
            region = '_'.join(parts[:-2])
        else:
            region = '_'.join(parts[:-1])
        
        if region not in region_groups:
            region_groups[region] = []
        region_groups[region].append(col)

# Find top 5 most important brain regions based on average feature importance
region_importance = {}
region_top_features = {}

for region, columns in region_groups.items():
    # Get feature indices for this region
    region_cols_indices = [list(X.columns).index(col) for col in columns]
    
    # Calculate region importance
    region_importance[region] = np.mean([importances[i] for i in region_cols_indices])
    
    # Sort features in this region by their importance
    region_features_with_importance = [
        (X.columns[idx], importances[idx]) 
        for idx in region_cols_indices
    ]
    
    # Sort features by importance in descending order
    sorted_region_features = sorted(
        region_features_with_importance, 
        key=lambda x: x[1], 
        reverse=True
    )
    
    # Store top 3 features for each region
    region_top_features[region] = [
        feature for feature, _ in sorted_region_features[:3]
    ]

# Print top 5 most important regions
top_regions = sorted(region_importance.items(), key=lambda x: x[1], reverse=True)[:20]
print("\nTop 5 most important brain regions based on average feature importance:")
for region, importance in top_regions:
    print(f"{region}: {importance:.4f}")

# Print top 3 features for each of these top regions
selected_features_global = []
print("\nTop 3 features for top regions:")
for region, importance in top_regions:
    print(f"\n{region} (Importance: {importance:.4f}):")
    top_features = region_top_features[region]
    for feature in top_features:
        selected_features_global.append(feature)
print(len(selected_features_global))
```

```python
def main(data_path, target_column, feature_columns=None, models_to_run=None, 
          test_size=0.3, use_adasyn=True, adasyn_neighbors=6, random_state=42):
    """
    Main function to run multiple classification models and compare their performance.
    
    Parameters:
    -----------
    data_path : str
        Path to the dataset file (Excel or CSV)
    target_column : str
        Column name of the target variable
    feature_columns : list, optional
        List of column names to use as features, if None will use all numeric columns
    models_to_run : list, optional
        List of model names to run. Options: 'xgboost', 'rf', 'dt', 'svm', 'nn', 'knn',
        'stack', 'bayes', 'gpc', 'gnb', 'tabnet', 'snn', 'att'.
        If None, all basic models will be run.
    test_size : float, default=0.3
        Proportion of data to use for testing
    use_adasyn : bool, default=True
        Whether to use ADASYN oversampling for imbalanced classes
    adasyn_neighbors : int, default=6
        Number of neighbors to use in ADASYN algorithm
    random_state : int, default=42
        Random seed for reproducibility
    
    Returns:
    --------
    dict
        Dictionary of model results
    pandas.DataFrame
        Comparison of model performance
    """
    # Initialize results dictionary
    models_results = {}
    
    # Set default models list if not provided
    if models_to_run is None:
        # By default, run only the basic models, not the advanced ones
        models_to_run = ['xgboost', 'rf', 'dt', 'svm', 'nn', 'knn', 'stack', 'gpc', 'gnb', 'tabnet', 'snn', 'att']
    else:
        models_to_run = [model.lower() for model in models_to_run]
    
    # Run XGBoost model
    if 'xgboost' in models_to_run:
        models_results['XGBoost'] = build_xgboost_model(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            random_state=random_state
        )
    
    # Run Random Forest model
    if 'rf' in models_to_run:
        models_results['RandomForest'] = build_random_forest_model(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            random_state=random_state
        )
    
    # Run SVM model
    if 'svm' in models_to_run:
        models_results['SVM'] = build_svm_model(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            random_state=random_state
        )
    
    # Run Neural Network model
    if 'nn' in models_to_run:
        models_results['NeuralNetwork'] = build_neural_network_model(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            random_state=random_state
        )
    
    # Run KNN model
    if 'knn' in models_to_run:
        models_results['KNN'] = build_knn_model(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            random_state=random_state
        )
    
    # Run Stacking Ensemble model
    if 'stack' in models_to_run:
        models_results['StackingEnsemble'] = build_stacking_ensemble(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            random_state=random_state
        )
    
    # Run Advanced Neural Network (Self-Normalizing Neural Network)
    if 'snn' in models_to_run:
        models_results['SelfNormalizingNN'] = build_advanced_neural_network(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            nn_type='snn',
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            model_name="SelfNormalizingNN",
            random_state=random_state
        )
    
    # Run Advanced Neural Network (Attention-based)
    if 'att' in models_to_run:
        models_results['AttentionNN'] = build_advanced_neural_network(
            data_path=data_path,
            target_column=target_column,
            feature_columns=feature_columns,
            nn_type='att',
            test_size=test_size,
            use_adasyn=use_adasyn,
            adasyn_neighbors=adasyn_neighbors,
            model_name="AttentionNN",
            random_state=random_state
        )
    
    # Compare models if more than one model was run
    if len(models_results) > 1:
        # Compare by accuracy
        comparison_acc = compare_models(models_results, metric='accuracy')
        print("\nModel Comparison (Accuracy):")
        print(comparison_acc)
        
        # Compare by ROC AUC
        comparison_roc = compare_models(models_results, metric='roc_auc')
        print("\nModel Comparison (ROC AUC):")
        print(comparison_roc)
        
        return models_results, comparison_acc
    elif len(models_results) == 1:
        print("\nOnly one model was run, no comparison performed.")
        return models_results, None
    else:
        print("\nNo models were run.")
        return {}, None
```
